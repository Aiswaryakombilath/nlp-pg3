```python
from transformers import pipeline
text_generator=pipeline("text-generation", model="gpt2")
summarizer=pipeline("summarization", model="facebook/bart-large-cnn")
sentiment_analyzer=pipeline("sentiment-analysis")
def show_menu():
  print("choose a task to perform")
  print("1. text generation")
  print("2.text summarization")
  print("3.sentiment analysis")
  print("4.exit")

input_text=input("enter the text: ")

while True:
  show_menu()
  choice=input("enter your choice(1-4):")

  if choice == "1":
    generated_text=text_generator(input_text, max_length=50, num_return_sequences=1)
    print("generated text:")
    print(generated_text[0]["generated_text"])
  elif choice == "2":
    summary=summarizer(input_text, max_length=150, min_length=30, do_sample=False)
    print("summary:")
    print(summary[0]["summary_text"])
  elif choice == "3":
    sentiment=sentiment_analyzer(input_text)
    print("===sentiment===")
    print(f"label:{sentiment[0]['label']}")
    print(f"score:{sentiment[0]['score']}")
  elif choice == "4":
    print("exiting")
    break
  else:
    print("invalid choice")
```
